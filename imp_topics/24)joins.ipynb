#### JOIN

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("field").getOrCreate()

data1 = [
  (1, "Arun",  "Chennai"),
  (2, "Meena", "Delhi"),
  (3, "Ravi",  "Mumbai"),
  (4, "Priya", "Hyderabad")
]

data2 =[
  (101, 1, "Laptop"),
  (102, 1, "Mouse"),
  (103, 3, "Speaker"),
  (104, 5, "Tablet")
]

df1 = spark.createDataFrame(data=data1,schema=["order_id","name","location"])
df2 = spark.createDataFrame(data=data2,schema=["item_id","order_id","item"])
```

#### Inner join

```python
df_inner = df1.join(df2,df1["order_id"]==df2["order_id"],how="inner")
df_inner.show()
```

### Left join

```python
df_left = df1.join(df2,df1["order_id"]==df2["order_id"],how="left")
df_left.show()
```

#### Right join

```python
df_right = df1.join(df2,df1["order_id"]==df2["order_id"],how="right")
df_right.show()
```

### Left anti

```python
df_left_anti = df1.join(df2,df1["order_id"]==df2["order_id"],how="anti")
df1.show()
df2.show()
df_left_anti.show()
```

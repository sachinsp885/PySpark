# DDL Schema

```python
my_ddl_schema ='''
                SaleID INT,
                Date STRING,
                CustomerName STRING,
                Product STRING,
                Quantity INT,
                Price INT
               '''
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("schema").getOrCreate()

df_ddl = spark.read.format('csv')\
                   .schema(my_ddl_schema)\
                   .option('Header',True)\
                   .load('C:/Git files/My git files/PySpark/files/sales_data.csv')
```

```python
df_ddl.printSchema()
```

# STRUCT TYPE 

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
my_struct_schema = StructType([
                                StructField('SaleID',IntegerType(),True),
                                StructField('Date',DateType(),True),
                                StructField('CustomerName',StringType(),True),
                                StructField('Product',StringType(),True),
                                StructField('Quantity',IntegerType(),True),
                                StructField('Price',IntegerType(),True)
])
```

```python
df_read_csv = spark.read.format('csv')\
                       .schema(my_struct_schema)\
                       .option("Header",True)\
                       .load('C:/Git files/My git files/PySpark/files/sales_data.csv')
```

```python
df_read_csv.printSchema()
```

### Null handling

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("date").getOrCreate()

df = spark.read.format("csv")\
               .option("inferSchema",True)\
               .option('header',True)\
               .load("C:/Git files/My git files/PySpark/files/test.csv")
```

# Dropping Null's

```python
#any -> drop the data from any column values
df = df.dropna('any')

#all -> only drop column where all the data having null
df.dropna('all')

#subset -> based column names
df = df.dropna(subset=['Last Name'])
```

# Fill Na

```python
#fillna -> method is used fill particular whereever null is there(for all the column)
df =df.fillna("Not available")
```

```python
df = df.fillna("Nun",subset=['Last Name'])
df.show()
```

```python
df = df.filter(col("age").isNotNull())
df = df.filter(col("age").isNull())
df.na.drop()
df.na.drop(subset=["name", "salary"])
df.na.fill({"salary": 0, "city": "Unknown"})
df.na.fill(0)
df.na.fill("NA")
df = df.withColumn("final_value",
                   coalesce(col("value1"), col("value2"), lit("NoData")))
df = df.withColumn(
    "status_clean",
    when(col("status").isNull(), "Missing").otherwise(col("status"))
)
```

# Split and indexing

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("date").getOrCreate()

df = spark.read.format("csv")\
               .option("inferSchema",True)\
               .option('header',True)\
               .load("C:/Git files/My git files/PySpark/files/test.csv")
```

```python
df.show()
```

##  Split

```python
# df.select(split("City",' ').alias('indexing')).show()
df.withColumn("City", split(col("City"), " ")).show()
```

# Indexing

```python
df = df.withColumn("Name_Email", split(col("Email"), "@")[0])
df.select("Name_Email").show()
```

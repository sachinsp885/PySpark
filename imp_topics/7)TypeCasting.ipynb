# Type Casting

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("select").getOrCreate()
df_csv_select = spark.read.format('csv')\
                          .option('inferSchema',True)\
                          .option('header',True)\
                          .load('C:/Git files/My git files/PySpark/files/sales_data.csv')
# C:\Git files\My git files\PySpark\files\sales_data.csv
```

```python
df_csv_select.printSchema()
```

```python
df_csv_select = df_csv_select.withColumn("Product",col("Product").cast(IntegerType()))
df_csv_select.printSchema()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("date").getOrCreate()

df = spark.read.format("csv")\
               .option("inferSchema",True)\
               .option('header',True)\
               .load("C:/Git files/My git files/PySpark/files/test.csv")
```

```python
df.show(10)
```

```python
df = df.withColumn("City",split("City"," "))
# df.show(10)
df_exp = df.withColumn("City",explode("City"))
df_exp.select("First Name","Last Name","City").show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *

spark = SparkSession.builder.appName("date").getOrCreate()

data = [
        ('A',1),
        ('A',2),
        ('A',3),
        ('B',1),
        ('B',2),
        ('B',1)
]

df = spark.createDataFrame(data=data,schema =['type','id'])
df.show()
```

```python
df1 = df.groupBy('Type').agg(collect_list('id').alias("id_list"))
df1.withColumn('id',explode(col('id_list'))).show()
df.show()
```
